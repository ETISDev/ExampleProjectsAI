# Area Plots

```python
df = pd.read_excel('./source_data/CanadaCitizenship.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)

df.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df['Total'] = df.sum(axis=1, numeric_only=True)
df.set_index('Country', inplace=True)
```

```python
df.sort_values(['Total'], ascending=False, axis=0, inplace=True)

# get the top 10 entries
df_top_10 = df.head()

years=list(range(1980,2014))
# transpose the dataframe
df_top_10 = df_top_10[years].transpose()

df_top_10.head()
```

```python
df_top_10.plot(kind='area',
               alpha = 0.3,
             stacked=False,
             figsize=(20,10))
plt.title('Immigration of top 5 countries')
plt.ylabel('No of immigrants')
plt.xlabel('Years')
plt.show()
```
